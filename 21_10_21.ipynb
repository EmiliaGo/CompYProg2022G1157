# Slicing (rebanado) de cadenas

Permite seleccionar subcadenas con e operador

[inicio : stop : incremento]

El stop no se incluye por eso debemos agregar una posicion extra

### Slicing y valores por defecto

inicio por defecto siempre ser√° 0 
stop: tama√±o del arreglo (indexado positivo)

Incremento: es 1 (indexado positivo)

```python
carrera = "Ingenier√≠a en computacion"
print (carrera [0:10:1] )
print (carrera [14:25:1] )
# lo mismo pero en indexado negativo ¬ø¬ø¬ø
print (carrera [-11:-1:1] )
print (carrera [:10:] )
print (carrera [14:25:] )

#solucion previa 
print ("------")
print (carrera [-11:-1:1] )
print (carrera [-11: :1] )
# toma un valor por defecto 


#ejercicio para incremento
print (carrera [0:10:2])
print (carrera[-1:-12:-1])
print (carrera[25:13:-1])
```

# Funci√≥n print

### Se emplea para desplegar informaci√≥n en la salida estandar. Normalmente un monitor

- Formas de usar print
- argumento por defecto de print (end)
- format ()
- operador f" "

```python
nombre = "Jos√©"
edad = 18
casado = False
estatura = 1.67
"""
Nombre : Jose 
edad:18
esta casado: False
estatura  en metros: 1.67
"""
# 1.-con comas, concatena haciendo un casting (conversion automanica a string)
# agrega espacios
print (nombre,edad,casado,estatura)
print ("Nombre: "+nombre +"\nEdad: "+ str(edad)+ "\nesta casado: "+ str(False)+ "\nEstatura en metros: "+ str(estatura))

# 2.- con el simbolo de "+"
# concatenacion en crudo 
# str (), para cambiarlo
# espacios " "
print ( nombre+" " +str(edad)+" " +str(casado)+" " +str(estatura) )

# 3.- dando formato a la salida con .format
print ("\nCon la funci√≥n .format")
print ("Nombre: {}\n\tEdad: {}\n\tEsta casado: {}\n\tEstatura en metros:{}".format (nombre,edad,casado,estatura))
# {} para orden 
print ("Nombre: {3}\n\tEdad: {2}\n\tEsta casado: {1}\n\tEstatura en metros:{0}".format (nombre,edad,casado,estatura))

# 4.- con el operador f" "
print ("\nCon el operador f\"\"") 
print (f"Nombre: {nombre}\n\tEdad: {edad}\n\tEsta casado: {casado}\n\tEstatura en metros:{estatura}")

# como manejar el argumento end
print ("\nComo usar el argumento end")
print (f"Hola {nombre}", end=" ")
print ( ", como estas?", end=" ")
```
